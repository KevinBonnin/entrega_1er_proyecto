Se procede a desanidar las columnas que estan anidadas

```python
def desanidar_dataframe(df, columna):
    def desanidar_dataframe(i):
        if isinstance(i, str):
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, i)
            return coincidencias
        elif isinstance(i, list):
            datos = []
            for e in i:
                if 'name' in e:
                    datos.append(e['name'])
            return datos
        elif isinstance(i, dict):
            if 'name' in i:
                return i['name']
        return None
    
    df[columna] = df[columna].apply(desanidar_dataframe)

    return df
```

```python
def desanidar_dataframe2(df, columna):
    def desanidar_dataframe2(i):
        if isinstance(i, str):
            pattern_name = r"'name': '([^']*)'"
            pattern_character = r"'character': '([^']*)'"
            coincidencias_name = re.findall(pattern_name, i)
            coincidencias_character = re.findall(pattern_character, i)
            return list(zip(coincidencias_name, coincidencias_character))
        elif isinstance(i, list):
            datos = []
            for e in i:
                if 'name' in e:
                    datos.append((e['name'], e.get('character')))
            return datos
        elif isinstance(i, dict):
            if 'name' in i:
                return (i['name'], i.get('character'))
        return None
    
    df[columna] = df[columna].apply(desanidar_dataframe2)

    return df
```

```python
def desanidar_dataframe3(df, columna):
    def desanidar_dataframe3(i):
        if isinstance(i, str):
            pattern = r"'job': 'Director', 'name': '([^']*)'"
            coincidencias = re.findall(pattern, i)
            return coincidencias
        elif isinstance(i, list):
            datos = []
            for e in i:
                if 'name' in e and e['job'] == 'Director':
                    datos.append(e['name'])
            return datos
        elif isinstance(i, dict):
            if 'name' in i and i['job'] == 'Director':
                return i['name']
        return None
    
    df[columna] = df[columna].apply(desanidar_dataframe3)

    return df
```
